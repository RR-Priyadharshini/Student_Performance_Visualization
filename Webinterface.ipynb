```python
# ------------------------------
# SECTION 1: STUDENT INTERFACE
# ------------------------------

import sqlite3
import pandas as pd
from IPython.display import display, HTML
import ipywidgets as widgets

# Create Database
conn = sqlite3.connect("study_logs.db")
cursor = conn.cursor()

# Create table
cursor.execute("""
CREATE TABLE IF NOT EXISTS study_logs(
    id INTEGER PRIMARY KEY AUTOINCREMENT,
        student_name TEXT,
            subject TEXT,
                hours INTEGER,
                    date TEXT
                    )
                    """)
conn.commit()

# ------------------------------
# Student Input Form
# ------------------------------

student_name = widgets.Text(description="Name:")
subject = widgets.Text(description="Subject:")
hours = widgets.IntText(description="Hours:")
date = widgets.Text(description="Date (YYYY-MM-DD):")

submit_button = widgets.Button(description="Save Log", button_style='success')

output_student = widgets.Output()

def save_study_log(a):
    with output_student:
        output_student.clear_output()
        cursor.execute("INSERT INTO study_logs(student_name,subject,hours,date) VALUES (?,?,?,?)",
                                       (student_name.value, subject.value, hours.value, date.value))
        conn.commit()
        print("Study log saved successfully!")
        show_data() # Refresh data after saving

submit_button.on_click(save_study_log)

display(HTML("<h2>üìò Student Interface - Enter Study Logs</h2>"))
display(student_name, subject, hours, date, submit_button, output_student)

# Display Updated Data
def show_data():
    df = pd.read_sql_query("SELECT * FROM study_logs", conn)
    with output_student: # Display data in the same output widget
        display(HTML("<h3>Current Study Logs:</h3>"))
        display(df)

show_data() # Initial display of data
```

```python
# ------------------------------
# SECTION 2: ADMIN PANEL
# ------------------------------

display(HTML("<h2>üîê Admin Login</h2>"))

admin_user = "admin"
admin_pass = "1234"

username = widgets.Text(description="Username:")
password = widgets.Password(description="Password:")
login_btn = widgets.Button(description="Login", button_style="info")
login_output = widgets.Output()

def admin_login(a):
    with login_output:
        login_output.clear_output()
        if username.value == admin_user and password.value == admin_pass:
            print("Login successful! Admin Panel Unlocked.")
            show_admin_panel()
        else:
            print("Incorrect username or password.")

login_btn.on_click(admin_login)

display(username, password, login_btn, login_output)


# ------------------------------
# Admin Panel (CRUD Operations)
# ------------------------------

def show_admin_panel():
    display(HTML("<h3>üõ† Admin Panel (CRUD Operations)</h3>"))

    # READ
    display(HTML("<h4>üìÑ Current Records</h4>"))
    df = pd.read_sql_query("SELECT * FROM study_logs", conn)
    display(df)

    # UPDATE widgets
    update_id = widgets.IntText(description="ID:")
    new_hours = widgets.IntText(description="New Hours:")
    update_btn = widgets.Button(description="Update", button_style='warning')
    update_output = widgets.Output()

    def update_record(a):
        with update_output:
            update_output.clear_output()
            cursor.execute("UPDATE study_logs SET hours=? WHERE id=?", (new_hours.value, update_id.value))
            conn.commit()
            print("Record updated successfully!")

    update_btn.on_click(update_record)

    display(HTML("<h4>‚úèÔ∏è Update Hours</h4>"))
    display(update_id, new_hours, update_btn, update_output)

    # DELETE widgets
    delete_id = widgets.IntText(description="ID:")
    delete_btn = widgets.Button(description="Delete", button_style='danger')
    delete_output = widgets.Output()

    def delete_record(a):
        with delete_output:
            delete_output.clear_output()
            cursor.execute("DELETE FROM study_logs WHERE id=?", (delete_id.value,))
            conn.commit()
            print("Record deleted successfully!")

    delete_btn.on_click(delete_record)

    display(HTML("<h4>üóë Delete Record</h4>"))
    display(delete_id, delete_btn, delete_output)

    # REFRESH
    display(HTML("<h4>üîÑ Updated Table</h4>"))
    df2 = pd.read_sql_query("SELECT * FROM study_logs", conn)
    display(df2)
```
