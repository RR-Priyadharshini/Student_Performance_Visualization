### **Section 1: Student Interface**


```python
!pip install ipywidgets
import sqlite3
import ipywidgets as widgets
from IPython.display import display, clear_output
```

```python
# Database setup
conn = sqlite3.connect("study_logs.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS study_logs (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    student_name TEXT,
    subject TEXT,
    hours INTEGER,
    date TEXT
)
""")
conn.commit()

"Database Ready!"
```

```python
# Student Interface ‚Äì Input Form
student_name = widgets.Text(description="Name:")
subject = widgets.Text(description="Subject:")
hours = widgets.IntText(description="Hours:")
study_date = widgets.Text(description="Date (YYYY-MM-DD):")

submit_btn = widgets.Button(description="Submit Log", button_style="success")
output1 = widgets.Output()

def submit_log(b):
    with output1:
        clear_output()
        cursor.execute("INSERT INTO study_logs (student_name, subject, hours, date) VALUES (?, ?, ?, ?)",
                       (student_name.value, subject.value, hours.value, study_date.value))
        conn.commit()
        print("‚úî Study log saved successfully!")

submit_btn.on_click(submit_log)

display(widgets.VBox([
    widgets.Label("üìò STUDENT INTERFACE"),
    student_name, subject, hours, study_date, submit_btn, output1
]))
```

### **Section 2: Admin Panel**

```python
# Admin Login
admin_user = "admin"
admin_pass = "1234"

username = widgets.Text(description="Admin:")
password = widgets.Password(description="Password:")
login_btn = widgets.Button(description="Login", button_style="warning")
login_output = widgets.Output()

def admin_login(b):
    with login_output:
        clear_output()
        if username.value == admin_user and password.value == admin_pass:
            print("‚úî Admin Login Successful!")
        else:
            print("‚ùå Wrong credentials")

login_btn.on_click(admin_login)

display(widgets.VBox([
    widgets.Label("üîê ADMIN LOGIN"),
    username, password, login_btn, login_output
]))
```

```python
# Admin ‚Äì View Records
admin_view_btn = widgets.Button(description="View All Logs", button_style="info")
admin_view_output = widgets.Output()

def admin_view(b):
    with admin_view_output:
        clear_output()
        cursor.execute("SELECT * FROM study_logs")
        rows = cursor.fetchall()
        if not rows:
            print("No records found.")
        else:
            for r in rows:
                print(r)

admin_view_btn.on_click(admin_view)

display(widgets.VBox([
    widgets.Label("üìÑ ADMIN PANEL ‚Äì VIEW RECORDS"),
    admin_view_btn, admin_view_output
]))
```

```python
# Admin ‚Äì Update Record
update_id = widgets.IntText(description="ID:")
new_hours = widgets.IntText(description="Hours:")
update_btn = widgets.Button(description="Update", button_style="primary")
update_output = widgets.Output()

def update_record(b):
    with update_output:
        clear_output()
        cursor.execute("UPDATE study_logs SET hours=? WHERE id=?", (new_hours.value, update_id.value))
        conn.commit()
        print("‚úî Record Updated Successfully!")

update_btn.on_click(update_record)

display(widgets.VBox([
    widgets.Label("‚úè UPDATE RECORD"),
    update_id, new_hours, update_btn, update_output
]))
```

```python
# Admin ‚Äì Delete Record
delete_id = widgets.IntText(description="ID to Delete:")
delete_btn = widgets.Button(description="Delete", button_style="danger")
delete_output = widgets.Output()

def delete_record(b):
    with delete_output:
        clear_output()
        cursor.execute("DELETE FROM study_logs WHERE id=?", (delete_id.value,))
        conn.commit()
        print("‚úî Record Deleted Successfully!")

delete_btn.on_click(delete_record)

display(widgets.VBox([
    widgets.Label("üóë DELETE RECORD"),
    delete_id, delete_btn, delete_output
]))
```
