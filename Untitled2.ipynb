```python
import sqlite3
import pandas as pd
from IPython.display import display
import ipywidgets as widgets

# Create database
conn = sqlite3.connect("study_logs.db")
cursor = conn.cursor()

# Create table (if not exists)
cursor.execute("""
CREATE TABLE IF NOT EXISTS study_logs (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    student_name TEXT,
    subject TEXT,
    hours INTEGER,
    date TEXT
)
""")

conn.commit()
```

```python
name = widgets.Text(description="Name:")
subject = widgets.Text(description="Subject:")
hours = widgets.IntText(description="Hours:")
date = widgets.Text(description="Date (YYYY-MM-DD):")

submit_btn = widgets.Button(description="Submit", button_style="success")
output = widgets.Output()

def save_data(b):
    with output:
        output.clear_output()
        cursor.execute("INSERT INTO study_logs (student_name, subject, hours, date) VALUES (?,?,?,?)",
                       (name.value, subject.value, hours.value, date.value))
        conn.commit()
        print("✔️ Data Saved Successfully!")

submit_btn.on_click(save_data)

widgets.VBox([name, subject, hours, date, submit_btn, output])
```

```python
df = pd.read_sql_query("SELECT * FROM study_logs", conn)
display(df)
```

```python
admin_user = "admin"
admin_pass = "1234"

username = widgets.Text(description="Username:")
password = widgets.Password(description="Password:")
login_btn = widgets.Button(description="Login", button_style="info")
login_out = widgets.Output()

is_logged_in = False

def admin_login(b):
    global is_logged_in
    with login_out:
        login_out.clear_output()
        if username.value == admin_user and password.value == admin_pass:
            is_logged_in = True
            print("✔️ Login Successful!")
        else:
            print("❌ Incorrect Credentials")

login_btn.on_click(admin_login)

widgets.VBox([username, password, login_btn, login_out])
```

```python
if is_logged_in:
    display(pd.read_sql_query("SELECT * FROM study_logs", conn))
else:
    print("⚠️ Please login to view admin panel.")
```

```python
update_id = widgets.IntText(description="ID:")
new_hours = widgets.IntText(description="New Hours:")
update_btn = widgets.Button(description="Update", button_style="warning")
update_out = widgets.Output()

def update_record(b):
    with update_out:
        update_out.clear_output()
        cursor.execute("UPDATE study_logs SET hours=? WHERE id=?", (new_hours.value, update_id.value))
        conn.commit()
        print("✔️ Record Updated!")

update_btn.on_click(update_record)
widgets.VBox([update_id, new_hours, update_btn, update_out])
```

```python
delete_id = widgets.IntText(description="Delete ID:")
delete_btn = widgets.Button(description="Delete", button_style="danger")
delete_out = widgets.Output()

def delete_record(b):
    with delete_out:
        delete_out.clear_output()
        cursor.execute("DELETE FROM study_logs WHERE id=?", (delete_id.value,))
        conn.commit()
        print("✔️ Record Deleted!")

delete_btn.on_click(delete_record)
widgets.VBox([delete_id, delete_btn, delete_out])
```
